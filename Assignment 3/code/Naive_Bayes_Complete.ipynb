```python
#Importing necessary libraries
from datasets import load_dataset
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
#Importing the huggingface dataset
data = load_dataset("dair-ai/emotion")
training_data = data["train"]
testing_data = data["test"]
validation_data = data['validation']
```

```python
#Vectorize the data
text_to_vector = CountVectorizer()
X_train = text_to_vector.fit_transform(training_data["text"])
X_test = text_to_vector.transform(testing_data["text"])
y_train = training_data["label"]
y_test = testing_data["label"]

#Fitting the data into model
alpha = 1
model = NaiveBayes()
model.fit(X_train, y_train,alpha)

#Getting the predictions of the model
predictions = model.predict(X_test)

#Determining the accuracy of the model
accuracy = model.evaluate_acc(y_test, predictions)
print("Accuracy of model : ",accuracy," %")
```
