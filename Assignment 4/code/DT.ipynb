# Decision Tree Model for Comparison

```python
from ucimlrepo import fetch_ucirepo
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import category_encoders as ce
import matplotlib.pyplot as plt
```

## Iris

```python
# fetch dataset 
iris = fetch_ucirepo(id=53) 
  
# data (as pandas dataframes) 
X = iris.data.features 
Y = iris.data.targets

# no of instances, no of features and classes
print(X.shape)
print(np.unique(Y))
```

```python
# DT model
DT = DecisionTreeClassifier(criterion='gini', splitter='best')

# train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
DT.fit(X_train, Y_train)
Y_pred = DT.predict(X_test)

print("Accuracy:", accuracy_score(Y_test, Y_pred))
```

```python
# get feature importances and plot them
feat_impt = DT.feature_importances_
i = range(len(feat_impt))
sorted_i = sorted(i, key=lambda i:feat_impt[i], reverse=True)

plt.figure(figsize=(8, 6))
plt.bar(range(len(feat_impt)), feat_impt[sorted_i])
plt.xticks(range(len(feat_impt)), sorted_i)
plt.xlabel('Feature')
plt.ylabel('Feature Importance')
plt.title('DecisionTreeClassifier Feature Importances')
plt.show()
```

## Poker Hand

```python
poker_train = pd.read_csv('datasets/poker-hand-training.csv')
poker_test = pd.read_csv('datasets/poker-hand-testing.csv')

val_train = poker_train.values
val_test = poker_test.values

X_train = poker_train.iloc[:, :10].values
Y_train = poker_train.iloc[:, 10].values
X_test = poker_test.iloc[:, :10].values
Y_test = poker_test.iloc[:, 10].values

# no of instances, no of features and classes
print(X_train.shape)
print(X_test.shape)
print(np.unique(Y_train))
print(Y_train[0])
```

```python
# DT model
DT = DecisionTreeClassifier(criterion='gini', splitter='best')

# train and test
X_train_sorted = np.sort(X_train, axis=1)
X_test_sorted = np.sort(X_test, axis=1)

DT.fit(X_train_sorted, Y_train)
Y_pred = DT.predict(X_test_sorted)

print("Accuracy:", accuracy_score(Y_test, Y_pred))
```
