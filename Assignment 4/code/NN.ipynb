# Neural Network Model for Comparison

```python
from ucimlrepo import fetch_ucirepo
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import category_encoders as ce
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelBinarizer
from sklearn.preprocessing import LabelEncoder
```

## Iris

```python
# fetch dataset 
iris = fetch_ucirepo(id=53) 
  
# data (as pandas dataframes) 
X = iris.data.features 
Y = iris.data.targets

# no of instances, no of features and classes
print(X.shape)
print(np.unique(Y))
```

```python
label_binarizer = LabelBinarizer()
Y_encoded = label_binarizer.fit_transform(Y)

# train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_encoded)

model = Sequential()
model.add(Dense(50, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(50, activation='relu'))
model.add(Dense(Y_train.shape[1], activation='softmax')) # The output layer should have as many neurons as there are classes

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, Y_train, epochs=200, batch_size=256)

# Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, Y_test)

print("Accuracy:", accuracy)
```

## Poker Hand

```python
# Load your data
poker_train = pd.read_csv('datasets/poker-hand-training.csv')
poker_test = pd.read_csv('datasets/poker-hand-testing.csv')

# Separate features and labels
X_train = poker_train.iloc[:, :10].values
Y_train = poker_train.iloc[:, 10].values
X_test = poker_test.iloc[:, :10].values
Y_test = poker_test.iloc[:, 10].values

# One-hot encode the labels
Y_train_encoded = to_categorical(Y_train)
Y_test_encoded = to_categorical(Y_test)

# no of instances, no of features and classes
print(X_train.shape)
print(np.unique(Y_train))
```

```python
# Define the model
model = Sequential()
model.add(Dense(50, activation='sigmoid', input_shape=(10,)))
model.add(Dense(50, activation='sigmoid'))
model.add(Dense(Y_train_encoded.shape[1], activation='softmax'))  # Output layer nodes = number of classes

# Compile the model with categorical_crossentropy
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, Y_train_encoded, epochs=200, batch_size=256)

# Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, Y_test_encoded)
print("Accuracy:", accuracy)
```

## Connect-4

```python
c4 = pd.read_csv('datasets/connect-4.csv')

c4_filtered = c4[c4.iloc[:, -1] != 'win']

val = c4_filtered.values
X = val[:, :42]
Y = val[:, 42]

# Print the shape of X and the unique values in Y
print(X.shape)
print(np.unique(Y))
print(X[0])
```

```python

def custom_encode(data):
    mapping = {'b': 0, 'x': 1, 'o': 2}
    return np.array([mapping[item] for item in data])

# Apply the custom encoding
X_encoded1 = np.array([custom_encode(row) for row in X])

label_binarizer = LabelBinarizer()
Y_encoded = label_binarizer.fit_transform(Y)

X_train, X_test, Y_train, Y_test = train_test_split(X_encoded1, Y_encoded, test_size=0.2)

model = Sequential()
model.add(Dense(50, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(50, activation='relu'))
model.add(Dense(Y_train.shape[1], activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, Y_train, epochs=200, batch_size=256)
loss, accuracy = model.evaluate(X_test, Y_test)

print("Accuracy:", accuracy)
```

## Image Segmentation

```python
image_seg_train = pd.read_csv('datasets/segmentation_train.csv')
image_seg_test = pd.read_csv('datasets/segmentation_test.csv')
  
val_train = image_seg_train.values
val_test = image_seg_test.values

X = np.concatenate((val_train[:, 1:20], val_test[:, 1:20]), axis=0)
Y = np.concatenate((val_train[:, 0], val_test[:, 0]), axis=0)

# no of instances, no of features and classes
print(X.shape)
print(np.unique(Y))

print(len(X[0]))
```

```python
label_binarizer = LabelBinarizer()
label_encoder = LabelEncoder()

Y_encoded = label_binarizer.fit_transform(Y)

X_df = pd.DataFrame(X)

X_encoded = X_df.apply(lambda col: label_encoder.fit_transform(col))

# train and test
X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y_encoded)

model = Sequential()
model.add(Dense(50, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(50, activation='relu'))
model.add(Dense(Y_train.shape[1], activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, Y_train, epochs=200, batch_size=256)

# Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, Y_test)

print("Accuracy:", accuracy)
```
